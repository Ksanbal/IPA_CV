```
import tensorflow as tf
```

```
a = tf.constant(3.0)
b = tf.constant(4.0)
c = tf.constant(5.0)

d = a*b+c
print(d)
# 그래프를 생성하는 과정
# TF 2.0에서는 바로 결과값 17이 나오게 된다
```

```
sess = tf.Session()
results = sess.run(d)
print(results)
```

```
w = tf.Variable(tf.random_normal(shape=[1], name='w'))
b = tf.Variable(tf.random_normal(shape=[1], name='b'))
x = tf.placeholder(tf.float32)

pred = x*w+b
```

```
y = tf.placeholder(tf.float32) # 정답을 넣을 placeholder
```

```
loss = tf.reduce_mean(tf.square(pred-y)) # 평균제곱오차
```

```
# 최적화 -> 오차를 줄이는 과정
optimizer = tf.train.GradientDescentOptimizer(0.01)
train_step = optimizer.minimize(loss)
```

```
x_train=[1,2,3,4]
y_train=[3,5,7,9]
```

```
sess = tf.Session()
results = sess.run(tf.global_variables_initializer()) # initializer가 랜덤값을 variables에 넣어준다
```

```
for i in range(1000):
  sess.run(train_step, feed_dict={x:x_train, y:y_train})
```

```
print(sess.run([w,b])) # 귀납법을 사용했기 때문에 딱 떨어지는 값이 아닌, 근사값을 구한것
```
